# feature engineering

```python
# converting name to title, surname and given name
df['Title']=" "
df['Surname'] =" "
df['given name']=" "
for i in range(df.shape[0]):
  df['Title'][i]=df['Name'][i].split(".")[0].split(",")[1].strip()
  df['Surname'][i]=df['Name'][i].split(".")[0].split(",")[0].strip()
  df['given name'][i]= df['Name'][i].split(".")[1].strip()
```

```python
for col in df.columns:
  if df[col].dtypes=="object":
    print(df[col].value_counts())
    
```

```python
# putting similar titles as same name
df['Title'] = df['Title'].replace(['Miss', 'Ms', 'Mlle', 'Lady', 'Mme', 'the Countess', 'Dona'], 
                                          'Ms')
df['Title'] = df['Title'].replace(['Dr', 'Col', 'Major', 'Jonkheer', 'Capt', 'Sir', 'Don', 'Rev'],
                                          'Dr/Military/Noble/Clergy')
```

* PassengerId is the unique id of the row and it doesn't have any effect on target
* Survived is the target variable we are trying to predict (0 or 1):
    * 1 = Survived
    * 0 = Not Survived
   
* Pclass (Passenger Class) is the socio-economic status of the passenger and it is a categorical ordinal feature which has 3  unique values (1, 2 or 3):
   * 1 = Upper Class
   * 2 = Middle Class
   * 3 = Lower Class
* Name, Sex and Age are self-explanatory
* SibSp is the total number of the passengers' siblings and spouse
* Parch is the total number of the passengers' parents and children
* Ticket is the ticket number of the passenger
* Fare is the passenger fare
* Cabin is the cabin number of the passenger
* Embarked is port of embarkation and it is a categorical feature which has 3 unique values (C, Q or S):
    * C = Cherbourg
    * Q = Queenstown
    * S = Southampton

```python

# converting age to kids,adults,middleage,old
def age_cat(x):
    if x < 18:
        return 'Kids'
    elif x >= 18 and x < 35:
        return 'Adults'
    elif x >= 35 and x<50:
        return 'middleage'
    elif x>=50:
        return 'old'    

# lets apply this Function
df['Age'] = df['Age'].apply(age_cat)

# lets check the Values
df['Age'].value_counts()
```

spliting data into features and target variables (x,y)

```python
x=df.drop(['Survived'], axis=1)
y=df['Survived'].to_frame()
```

```python
x
```

```python
y
```

normalization - same scale

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x = mms.fit_transform(x)
```

train test spliting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0 ) #20% will be test 
# random state=0 to get same values for train &test every time we run it
print('Shape of X_train is ', X_train.shape)
print('Shape of X_test is ', X_test.shape)
print('Shape of Y_train is ', Y_train.shape)
print('Shape of Y_test is ', Y_test.shape)
```

random forest classifier model

```python
from sklearn.ensemble import RandomForestClassifier 
model = RandomForestClassifier(n_estimators = 300 ,
                              #  min_samples_leaf = 5,
                              #  max_features = 0.5,# randomly select 50% of the columns
                              #  n_jobs = -1,#  use of all the cores
                              #  oob_score = True
                               )
model.fit(X_train , Y_train)
Y_pred=model.predict(X_test)
```

accuracy testing

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(Y_test,Y_pred)
confusion_matrix
```

```python
print(metrics.classification_report(Y_pred,Y_test))
```

```python
print(metrics.accuracy_score(Y_test,Y_pred))
```
