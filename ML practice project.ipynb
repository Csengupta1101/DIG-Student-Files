# RANDOM FOREST CLASSIFIER

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\data science asg nd notes\csv files\titanic.csv")
```

```python
df["Age"].fillna(df["Age"].mean(),inplace = True)
df["Embarked"].fillna(value = 'S', inplace= True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(["PassengerId","Cabin"], axis =1)
```

```python
sex = pd.get_dummies(df["Sex"])
em = pd.get_dummies(df["Embarked"])
```

```python
em
sex
```

```python
df = pd.concat([df, sex,em], axis = 1)
df
```

```python
df = df.drop(["Name", "Embarked","Sex", "Ticket"], axis=1)
df.head()
```

```python
df
```

```python
x = df[["Pclass","Age","SibSp","Parch","Fare","female","male","C","S"]]
y = df["Survived"]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2)
```

```python
rdfc = RandomForestClassifier(n_estimators = 1000)
rdfc
```

```python
rdfc= rdfc.fit(xtrain,ytrain)
```

```python
ypred = rdfc.predict(xtest)
```

```python
print("Accuracy:" , metrics.accuracy_score(ytest, ypred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, ypred)
cm
```
